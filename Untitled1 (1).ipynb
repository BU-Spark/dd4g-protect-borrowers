```python
from sqlalchemy import create_engine
import pandas as pd

db_url = "postgresql://ds701_update_read:tools4dsTeamATeamB@centerbeam.proxy.rlwy.net:34068/railway"

engine = create_engine(db_url)

table_names = pd.read_sql("""
    SELECT tablename
    FROM pg_catalog.pg_tables
    WHERE schemaname != 'pg_catalog' AND schemaname != 'information_schema'
""", engine)

display(table_names)
display(f'Number of tables found: {len(table_names):,}')

# --- Start of cases_df building logic, moved to this cell ---
case_frames = []

for table in table_names['tablename']:
    print(f"\n{'='*40}")
    print(f"TABLE: {table}")
    print('='*40)

    # Preview head of each table - now also including file_date
    # Assuming 'file_date' exists in all tables of interest for preview
    df_preview = pd.read_sql(f"SELECT *, file_date FROM {table} LIMIT 5", engine)
    display(df_preview)

    # Logic to build the full cases_df for analysis
    columns = pd.read_sql(
        f"""
        SELECT column_name
        FROM information_schema.columns
        WHERE table_name = '{table}'
        """,
        engine,
    )['column_name'].tolist()

    required_cols = {'case_id', 'court_name', 'plaintiff_name', 'file_date'}
    if required_cols.issubset(columns):
        query = f"SELECT case_id, court_name, plaintiff_name, file_date FROM {table}"
        df_full = pd.read_sql(query, engine)
        df_full['source_table'] = table
        case_frames.append(df_full)

if not case_frames:
    raise ValueError('No table contains case_id, court_name, plaintiff_name, and file_date.')

cases_df = pd.concat(case_frames, ignore_index=True).drop_duplicates()
cases_df['court_name'] = cases_df['court_name'].fillna('Unknown court')
cases_df['plaintiff_name'] = cases_df['plaintiff_name'].fillna('Unknown plaintiff')

print(f'Total rows used for EDA: {len(cases_df):,}')
display(f'Total rows in combined dataset: {len(cases_df):,}')
display(cases_df.head())

court_counts = (
    cases_df.groupby('court_name')['case_id']
    .nunique()
    .sort_values(ascending=False)
    .reset_index(name='case_count')
)

plaintiff_counts = (
    cases_df.groupby('plaintiff_name')['case_id']
    .nunique()
    .sort_values(ascending=False)
    .reset_index(name='case_count')
)

display(court_counts.head(50))
display(plaintiff_counts.head(50))

top_courts = court_counts.head(20)
top_plaintiffs = plaintiff_counts.head(50)
```

```python
import re
import pandas as pd

# Define the regex pattern to match the specified plaintiff names
# The pattern accounts for variations like 'GE Money Bank' and 'GE Capital Retail Bank'
# It also uses re.IGNORECASE for case-insensitive matching
regex_pattern = r'synchrony bank|ge capital retail bank|ge money bank'

# Filter the cases_df DataFrame based on the plaintiff_name column
# using the regex pattern. Using `na=False` to handle potential NaN values in plaintiff_name.
filtered_plaintiffs_df = cases_df[cases_df['plaintiff_name'].str.contains(regex_pattern, case=False, na=False)]

# Group by plaintiff_name and file_date to retain date information
plaintiffs_of_interest_counts = filtered_plaintiffs_df.groupby(['plaintiff_name', 'file_date'])['case_id'].nunique().reset_index(name='case_count')

# Convert 'file_date' to datetime format (YYYY-MM-DD)
plaintiffs_of_interest_counts['file_date'] = pd.to_datetime(plaintiffs_of_interest_counts['file_date'], format='%Y%m%d')

# Sort by case_count in descending order and then by file_date
plaintiffs_of_interest_counts = plaintiffs_of_interest_counts.sort_values(by=['case_count', 'file_date'], ascending=[False, True])

print(f"Number of cases for plaintiffs similar to 'Synchrony Bank', 'GE Capital Retail Bank', or 'GE Money Bank': {plaintiffs_of_interest_counts['case_count'].sum():,}")
display(plaintiffs_of_interest_counts)
display(f"Total cases for these plaintiffs: {plaintiffs_of_interest_counts['case_count'].sum():,}")
```

```python
import re

# Define a regex pattern that includes common misspellings for 'assignee' and 'successor'
# Common misspellings include: assingee, asignee, assinee, succesor, sucsessor, succesess, succesessors
comprehensive_pattern = r'ass|suc|nee'

# Filter plaintiffs containing either 'assignee' or 'successor' (including misspellings)
plaintiffs_with_either = plaintiffs_of_interest_counts[
    plaintiffs_of_interest_counts['plaintiff_name'].str.contains(comprehensive_pattern, case=False, na=False)
]

# Filter plaintiffs containing neither 'assignee' nor 'successor' (including misspellings)
plaintiffs_without_either = plaintiffs_of_interest_counts[
    ~plaintiffs_of_interest_counts['plaintiff_name'].str.contains(comprehensive_pattern, case=False, na=False)
]

print("\n--- Plaintiffs containing EITHER 'assignee' or 'successor' (including misspellings) ---")
display(plaintiffs_with_either)
display(f"Total cases for plaintiffs containing either 'assignee' or 'successor' (including misspellings): {plaintiffs_with_either['case_count'].sum():,}")

print("\n--- Plaintiffs containing NEITHER 'assignee' nor 'successor' (including misspellings) ---")
display(plaintiffs_without_either)
display(f"Total cases for plaintiffs containing neither 'assignee' nor 'successor' (including misspellings): {plaintiffs_without_either['case_count'].sum():,}")
```

### Subtask:

**Reasoning**:
The subtask requires calculating yearly case counts for `plaintiffs_with_either`, then filtering for years 2024 and 2025, and finally calculating and displaying monthly case counts for these specific years. This involves extracting year and month from the `file_date` column and performing aggregations.



```python
plaintiffs_with_either['year'] = plaintiffs_with_either['file_date'].dt.year

yearly_case_counts = plaintiffs_with_either.groupby('year')['case_count'].sum().reset_index()
print("\n--- Yearly Case Counts for Plaintiffs with 'Assignee' or 'Successor' ---")
display(yearly_case_counts)

filtered_2024_2025 = plaintiffs_with_either[plaintiffs_with_either['year'].isin([2024, 2025])].copy()
filtered_2024_2025['month'] = filtered_2024_2025['file_date'].dt.month

monthly_case_counts_2024_2025 = filtered_2024_2025.groupby(['year', 'month'])['case_count'].sum().reset_index()
monthly_case_counts_2024_2025['month_name'] = monthly_case_counts_2024_2025['month'].apply(lambda x: pd.to_datetime(str(x), format='%m').strftime('%B'))
monthly_case_counts_2024_2025 = monthly_case_counts_2024_2025.sort_values(by=['year', 'month'])

print("\n--- Monthly Case Counts for 2024 and 2025 (Plaintiffs with 'Assignee' or 'Successor') ---")
display(monthly_case_counts_2024_2025)
```
